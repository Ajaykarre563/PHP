# **What is OOP?**

Object-Oriented Programming (OOP) is a programming paradigm that organizes code into classes and objects.

Class: A blueprint for creating objects.

Object: An instance of a class.

# **Why use OOP?**

**OOP offers several benefits:**

Modularity: Code is easier to manage and debug.

Reusability: Use the same code (class) in multiple projects.

Scalability: Easier to scale large applications.

Security: Encapsulation protects data.

# **OOP with PHP â€“ Basic Example**

```
<?php
// Define a class
class Car {
    public $brand;

    function setBrand($name) {
        $this->brand = $name;
    }

    function getBrand() {
        return $this->brand;
    }
}

// Create an object
$myCar = new Car();
$myCar->setBrand("Toyota");

echo "My car brand is " . $myCar->getBrand();
?>

```

# **âœ… What is a Constructor in PHP?**

A constructor is a special method in a class that automatically gets called when a new object is created.
In PHP, itâ€™s defined using __construct().

## **ðŸ”¹ Why Only One Constructor?**

PHP supports only a single __construct() method per class. If you try to define multiple constructors, only the last one will be used â€” the previous ones will be overwritten.



```
<?php
class Student {
    public $name;
    public $age;

    // Constructor method
    function __construct($name, $age) {
        $this->name = $name;
        $this->age = $age;
        echo "Student Name: " . $this->name . "<br/>";
        echo "Age: " . $this->age . "<br/>";
    }
}

// Creating an object (constructor is called automatically)
$student1 = new Student("Adi", 20);
?>

```





```
Student Name: Adi
Age: 20

```

Answer: Inheritance is a feature in object-oriented programming (OOP) that allows one class (child/subclass) to inherit properties and methods from another class (parent/superclass). It promotes code reusability and helps in organizing code in a hierarchical manner.

# **2. Make Parent and Child Class**



```
<?php
// Parent class
class Animal {
    public $name;

    public function eat() {
        echo "$this->name is eating.";
    }
}

// Child class
class Dog extends Animal {
    public function bark() {
        echo "$this->name is barking.";
    }
}

// Create an object of Dog
$dog = new Dog();
$dog->name = "Buddy";
$dog->eat();  // Inherited method
echo "<br>";
$dog->bark(); // Child class method
?>

```

# **âœ… 1. What is an Access Modifier?**

Access Modifiers in PHP define the visibility or accessibility of class properties and methods. They control where and how class members can be accessed â€” from inside the class, outside, or from a child class.



# **âœ… 2. Types of Access Modifiers**
PHP supports three types of access modifiers:

public

protected

# **sample code**



```
<?php
class Person {
    public $name = "John";
    protected $age = 30;
    private $salary = 50000;

    public function showDetails() {
        echo "Name: $this->name, Age: $this->age, Salary: $this->salary";
    }
}

class Employee extends Person {
    public function showEmployeeAge() {
        echo "Age: $this->age"; // Allowed (protected)
    }
}

$person = new Person();
echo $person->name;         // âœ… public: accessible
// echo $person->age;       // âŒ protected: error
// echo $person->salary;    // âŒ private: error
echo "<br>";    

$person->showDetails();     // âœ… All visible within the class private
echo "<br>";
$emp = new Employee();
$emp->showEmployeeAge();    // âœ… Protected property accessed in subclass
?>

```

# **âœ… 1. What is Method Overriding?**
Method Overriding occurs when a child class defines a method with the same name and parameters as a method in its parent class.
The method in the child class replaces the parent version during execution.

# **ðŸ”¹ Example:**



```
<?php
class ParentClass {
    public function greet() {
        echo "Hello from Parent!";
    }
}

class ChildClass extends ParentClass {
    public function greet() {
        echo "Hello from Child!";
    }
}

$obj = new ChildClass();
$obj->greet(); // Output: Hello from Child!
?>

```



# **sample code**



```
<?php
class Teacher {
    public function nextExam() {
        echo "next exam is maths";
    }

    public function age() {
        echo "my age is 40";
    }
}

class Student extends Teacher {
    public function age() {
        echo "my age is 20";
    }
}

$st1 = new Student();
$st1->age();
?>
```

# **âœ… 2. What is Property Overriding?**

Property Overriding happens when a child class defines a property with the same name as one in the parent class.
The child class's property hides the parentâ€™s property.

# **sample code:**



```
<?php
class Teacher {
    public $city = "delhi"; // overriding the Property
    public function nextExam() {
        echo "next exam is maths";
    }

    public function age() {
        echo "my age is 40";
    }
}

class Student extends Teacher {
    public $city = "noida"; // Overriding the property
    public function age() {
        echo "my age is 20";
    }
}

$st1 = new Student();
$st1->age();
echo "<br>";
echo $st1->city; // Output: noida
?>
```

# **1. Use of final Keyword with Class**

When a class is declared as final, it cannot be extended (inherited) by any other class.



```
final class Animal {
    public function sound() {
        echo "Animal sound";
    }
}

// This will cause an error
class Dog extends Animal {
    public function sound() {
        echo "Dog barks";
    }
}

```



# **ðŸ”¹ 2. Use of final Keyword with Method**

When a method is declared as final, it cannot be overridden by any child class.



```
class Animal {
    final public function sound() {
        echo "Animal sound";
    }
}

class Dog extends Animal {
    // This will cause an error
    public function sound() {
        echo "Dog barks";
    }
}

```

The static keyword is used to declare class-level methods and properties.

They belong to the class itself, not to any object (instance).



# **ðŸ”¹ 2. When to Create Static Methods and Properties?**

When the method or property:

Doesn't depend on object state.

Performs a utility or shared task.

Should be accessible without creating an object.

# **ðŸ”¹ 3. Why Create Static Methods and Properties?**

Memory-efficient: No need to instantiate objects.

Useful for constants, counters, or utility functions.

Shared logic across all objects.

# **sample code**



```
<?php
class Honda {
    static public $countryName = "Japan";

    static function companyName() {
        echo "Honda";
    }
}

Honda::companyName();    // Calls the static method
echo "<br>";
echo Honda::$countryName; // Accesses the static property
?>

```

# **ðŸ”¹ What is Late Static Binding in PHP?**
Late Static Binding allows a static method in a parent class to call a static method or property from the child class at runtime.

It solves the problem of self:: being bound to the class where the method is defined, not the one that called it.

âœ… Late Static Binding uses static:: instead of self::.



```
<?php
class CountrySale {
    static public $totalSale = 1000;

    function getTotalSale() {
        echo static::$totalSale . "<br>";  // Late static binding to $totalSale
    }

    function getAreaName() {
        static::areaName();  // Late static binding to areaName()
    }

    static function areaName() {
        echo "India<br>";
    }
}

class CitySale extends CountrySale {
    static public $totalSale = 50;

    static function areaName() {
        echo "Noida<br>";
    }
}

// Create object of child class
$sale = new CitySale();
$sale->getTotalSale();   // Output: 50 (from CitySale)
$sale->getAreaName();    // Output: Noida (from CitySale)
?>

```

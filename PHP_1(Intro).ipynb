# **what is mean by PHP?**

PHP is a programming language used for web development.

It stands for Hypertext Preprocessor.

It is a server-side scripting language, meaning it runs on the server rather than the user's browser.

PHP is faster than other scripting languages like ASP and JSP.

# **What is a Server-Side Language?**

A server-side language is a programming language that runs on a web server instead of the user’s browser. It is used to process requests, interact with databases, and generate dynamic content before sending it to the user’s device.



# **How Does It Work?**

A user visits a website and makes a request (e.g., logging in).

The request is sent to the web server, where a server-side language (like PHP, Python, or Node.js) processes it.

The server retrieves data (e.g., user details) from a database if needed.

The server generates an HTML page and sends it to the user’s browser.

**What is Echo?**

echo is a PHP statement used to output text to a webpage.

**Use of Echo?**

It is used to display strings, variables, HTML content, etc.

**Case Insensitive:**

echo is not case-sensitive, so ECHO, Echo, and echo all work the same way.

**Line Break in Echo:**

```
You can add line breaks using <br> in HTML or \n in CLI output.

echo "Hello<br>World!";

```



# **Multiple echo in a Single Line in PHP**

echo to print multiple values in a single line by separating them with commas.



```
<?php
echo "Hello ", "Mr. ", "Ajay ", "Karre!";
?>

```



**Output**

```
Hello Mr. Ajay Karre!

```

```
<?php
// 1. String
$name = "John Doe";
echo "String: " . $name . "<br/>";

// 2. Integer
$age = 30;
echo "Integer: " . $age . "<br/>";

// 3. Float (or Double)
$salary = 45000.50;
echo "Float: " . $salary . "<br/>";

// 4. Boolean
$isEmployed = true;
echo "Boolean: " . ($isEmployed ? 'true' : 'false') . "<br/>";

// 5. Array
$colors = array("red", "green", "blue");
foreach ($colors as $color) {
    echo $color . "<br/>";
}

// 6. Object
class Car {
    public $brand;
    
    public function setBrand($brand) {
        $this->brand = $brand;
    }
    
    public function getBrand() {
        return $this->brand;
    }
}

$myCar = new Car();
$myCar->setBrand("Honda");
echo "Car Brand: " . $myCar->getBrand();
echo "<br/>";

// 7. NULL
$unknown = NULL;
echo "NULL: " . var_export($unknown, true) . "<br/>";


?>
```

**What is a Resource?**

A resource in PHP is a special variable that holds a reference to an external resource (like a database connection, file handle, or image).

It is not a data type like string or integer but rather a handle to an external operation.

# **Using echo with Inline HTML**



```
<?php
echo "<h1 style='color:green'>This is a heading</h1>";
echo "<p style='color:blue'>This is a paragraph.</p>";
?>

```



# **Using echo with String Concatenation**



```
<?php
$name = "John Doe";
echo "<h2>Welcome, " . $name . "!</h2>";
?>

```



# **Using PHP Inside HTML (<?php echo ?> or <?= ?> Shorthand)**



```
<h1>Hello, <?php echo $name; ?></h1>
<h2>Welcome, <?= $name; ?></h2>  <!-- Shorthand for echo -->

```



## **Using print (Similar to echo)**



```
<?php
print "<h3>This is an h3 tag</h3>";
?>

```

# **What are Constants?**

A constant is a value that cannot be changed after it has been defined.

Constants are global and can be accessed anywhere in the script.

They are usually written in uppercase by convention.



Screenshot 2025-03-25 143437.png

# **Sample code**



```
<?php
const DATA1 = "Ajay Karre";
const DATA2 = "peter";
echo DATA1;
echo "<br/>";
 // Output: anil sidhu
?>

<?php
$name = "bruce";
$name = "tony";
echo $name;  // Output: tony
?>

```

# **Two Ways to Define a Constant in PHP**

**Method 1: Using define()**



```
define("SITE_NAME", "MyWebsite");
echo SITE_NAME;  // Output: MyWebsite

```



# **Method 2: Using const**



```
const SITE_NAME = "MyWebsite";
echo SITE_NAME;  // Output: MyWebsite

```

# **Interview Questions and Answers on Constants in PHP**

What is a constant in PHP, and how is it different from a variable?

→ A constant is a fixed value that cannot be changed once defined, whereas a variable can be reassigned.

How do you define a constant in PHP? Explain define() vs const.

→ Use define("NAME", "value") (works inside functions) or const NAME = "value" (only in global scope).

Can we change the value of a constant in PHP? Why or why not?

→ No, a constant cannot be changed after declaration because it is immutable.

Are constants case-sensitive? How can we make them case-insensitive?

→ Yes, constants are case-sensitive by default. Use define("NAME", "value", true); (deprecated in PHP 7.3+).

What happens if you try to redefine a constant?

→ It causes a fatal error: "Cannot redeclare constant 'NAME'".

# **What is a data type**?

A data type defines the type of value a variable can store in a programming language.

Total data types in PHP:

**PHP has 8 main data types.**

Names of data types in PHP:

Scalar Types: Integer, Float, String, Boolean

Compound Types: Array, Object

Special Types: NULL, Resource



```
<?php
$name = "Anil Sidhu";
echo var_dump($name);   // String data type
echo "<br/>";

$num = 100;  
echo var_dump($num);    // Integer data type
echo "<br/>";

$numFloat = 10.0;  
var_dump($numFloat); // Float (double) data type
echo "<br/>";

$bool = true;  
echo var_dump($bool);   // Boolean data type
echo "<br/>";

$arr = ["anil", "sam", 10];  
echo var_dump($arr);    // Array data type
echo "<br/>";


$null = NULL;  
echo var_dump($null);   // Null data type
?>
```

# **Comments in PHP**

# **What is a Comment in the Code?**
A comment is a line in the code that is ignored by the PHP interpreter. It is used to explain code, make notes, or temporarily disable code execution.

# **Types of Comments in PHP:**

PHP supports three types of comments:

Single-line comment (//)

Single-line comment (#)

Multi-line comment (/* */)

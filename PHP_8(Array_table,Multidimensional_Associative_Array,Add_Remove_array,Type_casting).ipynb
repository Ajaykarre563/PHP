# **Array Display in Table Format**

**"Array Display in Table" means taking data stored in an array and representing it in a structured HTML table format.**





```
<?php
$users = [
    [1, "Anil", "Noida", "anil@test.com"],
    [2, "Sam", "Delhi", "sam@test.com"],
    [3, "Peter", "Gurgaon", "peter@test.com"]
];
?>

<table border="1">
    <tr>
        <th>ID</th>
        <th>Name</th>
        <th>City</th>
        <th>Email</th>
    </tr>
    <?php
    for ($i = 0; $i < count($users); $i++) {
        echo "<tr>";
        for ($j = 0; $j < count($users[$i]); $j++) {
            echo "<td>" . $users[$i][$j] . "</td>";
        }
        echo "</tr>";
    }
    ?>
</table>

```

# **Multidimensional Associative Array in PHP**

A multidimensional associative array is an array that contains other arrays, where keys are named (associative) instead of being indexed numerically. This helps in organizing data in a structured way, making it easier to retrieve values.

# **1. What is a Multidimensional Associative Array?**

A multidimensional associative array is an array of arrays where each inner array uses named keys instead of numerical indexes.

**Regular Indexed Array (Numerical Keys):**



```
$users = [
    [1, "Anil", "Noida", "anil@test.com"],
    [2, "Sam", "Delhi", "sam@test.com"]
];

```



**ðŸ”¹ Associative Array (Named Keys):**



```
$users = [
    "user1" => ["id" => 1, "name" => "Anil", "city" => "Noida", "email" => "anil@test.com"],
    "user2" => ["id" => 2, "name" => "Sam", "city" => "Delhi", "email" => "sam@test.com"]
];

```



# **2. Use of Multidimensional Associative Array**

âœ” **Easy Data Retrieval** â€“ Instead of remembering numeric indexes, we can use meaningful keys.

âœ” **Use in APIs & JSON** â€“ Often used in APIs when returning JSON data.



```
echo $users["user1"]["name"];  // Output: Anil

```

# **Remove & Add Elements in an Array (PHP)**

In PHP, arrays are dynamic, meaning we can add and remove elements easily. Let's go step by step with examples.


**Example: Using array_push()**



```
$fruits = ["Apple", "Banana"];
array_push($fruits, "Mango");  // Add "Mango"
print_r($fruits);

```



# **Output:**



```
Array ( [0] => Apple [1] => Banana [2] => Mango )

```



**âœ… Example: Using Direct Assignment**



```
$fruits[] = "Orange";  // Adds "Orange" at the next index
print_r($fruits);

```



# **Sample code**



```
<?php
$fruits = ["Apple", "Banana"];

array_push($fruits, "Mango");  // Add "Mango"
print_r($fruits);
echo "<br>";  // Add a line break for readability

$fruits[] = "Orange";  // Adds "Orange" at the next index
print_r($fruits);
?>

```

# **Add Multiple Elements**

Use array_push() to add multiple elements at once.

# **Example**



```
$fruits = ["Apple", "Banana"];
array_push($fruits, "Mango", "Grapes");
print_r($fruits);

```



# **Output**



```
Array ( [0] => Apple [1] => Banana [2] => Mango [3] => Grapes )

```



# **You can also merge arrays using array_merge():**



```
$more_fruits = ["Pineapple", "Watermelon"];
$fruits = array_merge($fruits, $more_fruits);
print_r($fruits);

```



# **3. Remove a Single Element**

Use unset() to remove an element by index.



```
$fruits = ["Apple", "Banana", "Mango"];
unset($fruits[1]);  // Removes "Banana"
print_r($fruits);

```



# **output:**



```
Array ( [0] => Apple [2] => Mango )

```



**ðŸ”¹ Notice that index 1 is missing but the other elements remain intact.**

To reindex the array, use:





```
$fruits = array_values($fruits);

```



# **sample code**



```
<?php
$fruits = ["Apple", "Banana", "Mango"];

unset($fruits[1]);  // Removes "Banana"
print_r($fruits);
echo "<br>";  // Line break for better readability

$fruits = array_values($fruits);  // Reindex the array
print_r($fruits);
?>

```

# **Type Casting in PHP**

# **1. What is Type Casting?**

Type casting in PHP is the process of converting a variable from one data type to another manually. It is useful when dealing with dynamic data types.

# **Type and Type casting are same...**

# **2. Types of Casting in PHP**

**PHP supports the following type conversions:**

(int) or (integer) â†’ Convert to an integer

(bool) or (boolean) â†’ Convert to a boolean

(float), (double), or (real) â†’ Convert to a floating point number

(string) â†’ Convert to a string

(array) â†’ Convert to an array

(object) â†’ Convert to an object

(unset) â†’ Convert to NULL

# **3. Syntax**

The general syntax of type casting is:



```
$new_variable = (new_type) $old_variable;

```

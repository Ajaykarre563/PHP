# **ðŸ”¹ What are Magic Methods?**

Magic methods in PHP are special methods that start with double underscores (__) and are automatically triggered by specific actions or behaviors in a class.

# **ðŸ”¹ What is the use of __invoke method?**

The __invoke() method in PHP is triggered when an object is used as a function. It allows an object to behave like a callable function.



# **sample code:**



```
<?php
class Calculator {
    public function __invoke($a, $b) {
        return $a + $b;
    }
}

$add = new Calculator();
echo $add(10, 20);  // Output: 30
?>

```

# **ðŸ”¹ What is the use of __get method?**

The __get($name) magic method is automatically called when trying to access a non-existent or private property of a class.



# **sample code**



```
<?php
class User {
    public $name = "Anil sidhu";
    private $age = 30;

    function __get($property){
        echo "$property property is not exist";
    }
}

$user = new User();
echo $user->name;       // Accesses public property â†’ Output: Anil sidhu
echo "<br>";
echo $user->age;        // Tries to access private property â†’ Triggers __get()
?>
```

# **ðŸ”¹ What is the use of __set method?**

The __set($name, $value) method is automatically called when writing to a private or non-existing property of an object.

# **sample code:**



```
<?php
class User {
    private $name;

    function __set($property, $value){
        // these function is used to set the value of private property
        // if property is not exist then it will be called
        
        echo "$property property can not set with $value value";
    }
}

$user = new User();
$user->name = "Anil sidhu";  // Triggers __set()
?>

```

# **ðŸ” Why use __set?**

It gives you control over how inaccessible or undefined properties are handled, like validation, logging, or dynamic assignment.

# **ðŸ”¹ What is the use of __call method?**

The __call($name, $arguments) method is triggered when invoking inaccessible or undefined methods in an object.

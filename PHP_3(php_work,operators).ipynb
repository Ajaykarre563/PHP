# **What are Operators?**

Operators are symbols that perform operations on variables and values. They are fundamental in programming languages for performing calculations, comparisons, and logical operations.

# **Example of Operators**

Examples of operators in programming:

Arithmetic Operators: +, -, *, /, %

Comparison Operators: ==, !=, >, <, >=, <=

Logical Operators: &&, ||, !

Assignment Operators: =, +=, -=, *=, /=

Bitwise Operators: &, |, ^, ~, <<, >>

# **Types of Operators**

Operators are categorized based on their functionality:

Arithmetic Operators (Used for mathematical calculations)

Relational (Comparison) Operators (Used to compare values)

Logical Operators (Used for logical conditions)

Assignment Operators (Used to assign values to variables)

Bitwise Operators (Used for bit-level operations)

Ternary Operator (A short form of if-else: condition ? value1 : value2)

# **Interview Questions and Answers on Operators**

**What is the difference between == and === in JavaScript?**

== checks for value equality with type conversion, while === checks for both value and type equality.

**How does the ternary operator work in Python?**

The ternary operator in Python follows the format: value_if_true if condition else value_if_false.

**Explain the use of bitwise operators with an example.**

Bitwise operators perform operations on binary numbers, e.g., 5 & 3 (binary AND) results in 1.

**How do logical operators work in conditional statements?**

Logical operators (and, or, not) evaluate boolean expressions to control the flow of conditions in statements.

# **arithmetic operators**



```
<?php
// Define two numbers
$a = 10;
$b = 3;

// Arithmetic operations
$addition = $a + $b;        // Addition
$subtraction = $a - $b;     // Subtraction
$multiplication = $a * $b;  // Multiplication
$division = $a / $b;        // Division (floating-point)
$modulus = $a % $b;         // Modulus (remainder)
$exponentiation = $a ** $b; // Exponentiation (a^b)
$floor_division = intdiv($a, $b); // Floor Division (integer division)

// Display results
echo "Addition: $a + $b = $addition\n";
echo "<br/>";
echo "Subtraction: $a - $b = $subtraction\n";
echo "<br/>";
echo "Multiplication: $a * $b = $multiplication\n";
echo "<br/>";
echo "Division: $a / $b = $division\n";
echo "<br/>";
echo "Modulus: $a % $b = $modulus\n";
echo "<br/>";
echo "Exponentiation: $a ** $b = $exponentiation\n";
echo "<br/>";
echo "Floor Division: intdiv($a, $b) = $floor_division\n";
echo "<br/>";
?>

```

## **Increment and Decrement Operators in PHP**

In PHP, increment (++) and decrement (--) operators are used to increase or decrease a variable’s value by 1. These operators can be used in pre or post forms.

# **Increment Operator (++)**

Pre-increment (++$var) → Increments the value first, then returns it.

Post-increment ($var++) → Returns the value first, then increments it.

# **Decrement Operator (--)**

Pre-decrement (--$var) → Decrements the value first, then returns it.

Post-decrement ($var--) → Returns the value first, then decrements it.



```
<?php
// Demonstrating Increment Operators
$x = 5; // Initial value

echo "<br>";
echo ++$x;
// Pre-increment: $x becomes 6, then prints 6

echo "<br>";
echo $x++;
// Post-increment: Prints 6, then $x becomes 7

echo "<br>";
echo $x;
// Now $x is 7, so it prints 7

echo "<br><br>"; // Adding space for clarity

// Demonstrating Decrement Operators
$y = 5; // Initial value

echo --$y;
// Pre-decrement: $y becomes 4 first, then prints 4

echo "<br>";
echo $y--;
// Post-decrement: Prints 4, then $y becomes 3

echo "<br>";
echo $y;
// Now $y is 3, so it prints 3
?>

```

Effect of increment and decrement on strings in PHP: ++ increases alphabetical characters ("a" becomes "b", "z" becomes "aa"), while -- is undefined for strings.

Effect of ++ on boolean or null: ++ on true keeps it true, on false changes it to 1, and on null changes it to 1.

Can ++ and -- be used on arrays? No, using ++ or -- on arrays results in a warning and does not modify the array.




# **Strings and String Operators in PHP**

**A string is a collection of characters enclosed in single (') or double (") quotes.**



```
$str1 = "Hello, World!"; // Double quotes
$str2 = 'PHP is awesome!'; // Single quotes

echo $str1;
echo "<br>";
echo $str2;

```



# **Difference between Single and Double Quotes**

Single quotes (') → Treats everything inside as plain text.

Double quotes (") → Allows variable interpolation and escape sequences.



```
$name = "John";
echo 'Hello, $name'; // Output: Hello, $name (No interpolation)
echo "<br>";
echo "Hello, $name"; // Output: Hello, John (Variable interpolated)

```



# **Using Variables Inside a String**

You can insert variables inside a string using double quotes or concatenation (.).



```
$city = "Hyderabad";
echo "I live in $city"; // Variable inside double quotes
echo "<br>";
echo 'I live in ' . $city; // Using concatenation

```

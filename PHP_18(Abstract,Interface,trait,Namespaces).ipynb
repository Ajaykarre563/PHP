# **ðŸ”¹ What is an Abstract Class in PHP?**

An abstract class in PHP is a class that cannot be instantiated directly. It is used as a base class and may contain abstract methods (methods without a body) that must be implemented in child classes.



# **ðŸ”¹ Why Do We Use Abstract Classes in PHP?**

To define a blueprint for other classes.

To ensure certain methods are mandatory in derived classes.

To promote code reusability and structure.



```
<?php
abstract class Animal {
    // Abstract method (no implementation here)
    abstract public function makeSound();

    // Concrete method
    public function eat() {
        echo "Eating...\n";
    }
}

class Dog extends Animal {
    public function makeSound() {
        echo "Bark!\n";
    }
}

$dog = new Dog();
$dog->makeSound(); // Output: Bark!
$dog->eat();       // Output: Eating...
?>

```

```
<?php
abstract class ProductFeatures {
    abstract function productDetails();
    abstract function productImages();
    abstract function productOwnerDetails(); // fixed this line
}

class UploadProduct extends ProductFeatures {
    function productDetails() {
        echo "product details";
    }

    function productImages() {
        echo "product images";
    }

    function productOwnerDetails() {
        echo "product owner";
    }
}

$upload = new UploadProduct();
$upload->productDetails();
echo "<br>";
$upload->productImages();
echo "<br>";
$upload->productOwnerDetails();
?>
```

# **ðŸ”¹ What is an Interface in PHP?**

An interface in PHP defines a contract for classes. It only contains method declarations, and no method bodies. Any class that implements an interface must define all its methods.



```
interface Logger {
    public function log($message);
}

```



# **ðŸ”¹ Why Do We Use Interfaces?**
To enforce method consistency across multiple classes.

To implement multiple inheritance (since PHP does not support multiple class inheritance).

To promote polymorphism and loose coupling in code design.



image.png



```
<?php
interface PaymentGateway {
    public function pay($amount);
}

class PayPal implements PaymentGateway {
    public function pay($amount) {
        echo "Paid â‚¹$amount using PayPal";
    }
}

$payment = new PayPal();
$payment->pay(500);
?>

```

**Traits are a mechanism for code reuse in PHP. They allow you to include methods in multiple classes without using inheritance. Traits help when:**

**You want to share methods across unrelated classes.**

You want to avoid code duplication.



# **ðŸ”¹ Make Traits and Class**
A trait is created using the trait keyword. Hereâ€™s a basic example:



```
trait Logger {
    public function log($message) {
        echo "Log message: $message";
    }
}

class User {
    use Logger;
}

```






# **sample code:**



```
<?php

trait parentCompany1 {
    function getTotalEmp() {
        echo 500;
    }

    function getTotalProject() {
        echo 35;
    }
}

trait parentCompany2 {
    function getTotalOffice() {
        echo 10;
    }
}

class Company {
    use parentCompany1;
    use parentCompany2;
}

// Create an object
$obj = new Company();

// Call trait methods
$obj->getTotalEmp();       // Output: 500
echo "<br>";
$obj->getTotalProject();   // Output: 35
echo "<br>";
$obj->getTotalOffice();    // Output: 10

```

# **sample2 code:**



```
<?php

trait ParentCompany1 {
    function getEmp() {
        echo "Employee Count from ParentCompany1: 200<br/>";
    }
}

trait ParentCompany2 {
    function getEmp() {
        echo "Employee Count from ParentCompany2: 300<br/>";
    }
}

class Company {
    use ParentCompany1, ParentCompany2 {
        ParentCompany1::getEmp insteadof ParentCompany2; // Use getEmp from ParentCompany1
        ParentCompany2::getEmp as getEmpCompany2; // Alias getEmp from ParentCompany2 because it has the same name
    }
}

$c1 = new Company();
$c1->getEmp();           // Output: from ParentCompany1
$c1->getEmpCompany2();   // Output: from ParentCompany2

?>

```

# **âœ… What is Namespace?**
Namespace is like a folder for your classes, functions, and constants to avoid name conflicts in large PHP projects.

Think of it like organizing files in folders so that two files with the same name donâ€™t clash.

# **âœ… Make Files and Folders for Namespace**
Create a folder named namespaces/ and inside it:

teacher.php

student.php

management.php

Main file: namespaces.php (optional: acts as a test runner)

# **âœ… Make Class and Namespace**
In teacher.php:



```
<?php
namespace teacher;

class JoiningDetails {
    public function joiningDate() {
        echo "Teacher Joining Date: 10-06-2020<br>";
    }
}
?>

```





In student.php:



```
<?php
namespace student;

class JoiningDetails {
    public function admissionDate() {
        echo "Student Admission Date: 01-01-2022<br>";
    }
}
?>

```



âœ… Example (in management.php)


```
<?php
include('teacher.php');
include('student.php');

$teacher = new teacher\JoiningDetails();
$teacher->joiningDate();

$student = new student\JoiningDetails();
$student->admissionDate();
?>

```

# **What is a Conditional Statement?**

A conditional statement is used to perform different actions based on whether a specified condition evaluates to true or false.

# **Types of Conditional Statements in PHP**

PHP supports the following conditional statements:

if Statement

if-else Statement

if-elseif-else Statement

switch Statement

# **What is the if Condition?**

The if statement is used to execute a block of code only if a specified condition evaluates to true.





```
if (condition) {
    // Code to execute if the condition is true
}

```

# **What happens if there is no break in a switch statement?**

If there is no break, PHP will execute all cases below the matching one until it finds a break or reaches the end of the switch block. This is called fall-through behavior.

# **Sample code**



```
<?php
$age = 10;          // Variable $age is assigned a value of 10
$name = "Anil";     // Variable $name is assigned a string value "Anil"

if ($age < 18) {    // Checks if age is less than 18
    echo "Display Kids content";  // Prints message if condition is true
}

echo "<br/>";  // Adds a line break in HTML output

if ($name == "Anil") {  // Checks if $name is "Anil"
    echo "<h1>Welcome, Anil Sidhu</h1>";  // Prints a welcome message in H1 heading
}

echo "Other";  // This will always be printed because it is outside the if conditions
?>

```

# **Logical Operators**

Logical operators in PHP are used to perform logical operations, typically in conditional statements. Here’s a breakdown:

Use of Logical Operators
Logical operators help in decision-making by combining multiple conditions.

Screenshot 2025-03-26 152727.png



```
<?php
$x = true;
$y = false;

// AND operator (and)
echo "AND (and): ";
var_dump($x and $y); // False
echo "<br>";

// OR operator (or)
echo "OR (or): ";
var_dump($x or $y); // True
echo "<br>";

// XOR operator (xor)
echo "XOR (xor): ";
var_dump($x xor $y); // True (only one should be true)
echo "<br>";

// AND operator (&&)
echo "AND (&&): ";
var_dump($x && $y); // False
echo "<br>";

// OR operator (||)
echo "OR (||): ";
var_dump($x || $y); // True
echo "<br>";

// NOT operator (!)
echo "NOT (!): ";
var_dump(!$x); // False (negates $x)
echo "<br>";
?>

```



Screenshot 2025-03-26 153216.png

# **Note**

**1.Logical operators in PHP are used to perform logical operations like AND, OR, NOT, and XOR to evaluate conditions.**

**2.&& has higher precedence than and, meaning && executes before assignment, while and has lower precedence.**

**3.|| has higher precedence than or, so || executes before assignment, whereas or is evaluated later.**

**4.xor returns true if only one of the conditions is true, e.g., true xor false gives true, but true xor true gives false.**

**5.! negates a condition, meaning !true becomes false, and !false becomes true.**










# **Loops in PHP**

**Loops in PHP are used to execute a block of code multiple times until a specified condition is met.**

**Types of Loops in PHP**

for Loop – Used when the number of iterations is known.

while Loop – Repeats as long as the condition is true.

do-while Loop – Executes at least once, then repeats if the condition is true.

foreach Loop – Iterates over an array or object.



# **for Loop**

✅ Used when the number of iterations is fixed.



```
<?php
for ($i = 1; $i <= 5; $i++) {
    echo "Iteration: $i <br>";
}
?>

```



# **while Loop**

# **✅ Runs as long as the condition remains true.**



```
<?php
$i = 1;
while ($i <= 5) {
    echo "Iteration: $i <br>";
    $i++;
}
?>

```



# **do-while Loop**

**✅ Executes the block at least once, even if the condition is false.**



```
<?php
$i = 6;
do {
    echo "Iteration: $i <br>";
    $i++;
} while ($i <= 5);
?>

```



# **foreach Loop (For Arrays)**

✅ Specifically used for iterating over arrays.



```
<?php
$colors = ["Red", "Green", "Blue"];
foreach ($colors as $color) {
    echo "Color: $color <br>";
}
?>

```

# **What is a break in a Loop?**

The break statement in PHP is used to immediately exit a loop



```
<?php
$table = 123;
for($start = 1; $start <= 10; $start++) {
    $result = $table * $start;
    if($result > 500) {
        break; // Exit the loop if the result exceeds 500
    }
    echo $table . " * " . $start . " = " . $result . "<br>";
}
?>
```



# **What is a continue in a Loop?**

The continue statement in PHP is used to skip the rest of the current iteration of a loop and move to the next iteration.



```
<?php
$table = 123;
for($start = 1; $start <= 10; $start++) {
    if($start % 2 != 0) { // If $start is odd
        continue; // Skip this iteration
    }
    echo $table . " * " . $start . " = " . ($table * $start) . "<br>";
}
?>
```





```
123 * 2 = 246
123 * 4 = 492
123 * 6 = 738
123 * 8 = 984
123 * 10 = 1230
```

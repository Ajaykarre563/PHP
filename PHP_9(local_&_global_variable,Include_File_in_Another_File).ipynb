# **Local and Global Variables in PHP?**

# **1. What is a Local Variable?**

A local variable is a variable declared inside a function and can only be accessed within that function. It does not exist outside the function.



```
<?php
function myFunction() {
    $x = 10; // Local variable
    echo "Value inside function: $x"; // Output: 10
}
myFunction();
echo $x; // ERROR: Undefined variable
?>

```





```
‚úÖ Output: Value inside function: 10
‚ùå Error: $x is not accessible outside the function.
```



# **2. What is a Global Variable?**

A global variable is declared outside any function and can be accessed anywhere except inside functions unless explicitly declared as global.



```
<?php
$y = 20; // Global variable

function myFunction() {
    echo $y; // ERROR: Undefined variable
}

myFunction();
?>

```





```
‚ùå Error: $y is not accessible inside the function.


```



# **3. How to Get a Global Variable Inside a Function?**

To use a global variable inside a function, use the global keyword or $GLOBALS array.

# **‚úÖ Method 1: Using global Keyword**



```
<?php
$y = 20; // Global variable

function myFunction() {
    global $y;  // Making $y accessible inside function
    echo "Value inside function: $y";
}

myFunction(); // Output: Value inside function: 20
?>

```



# **‚úÖ Method 2: Using $GLOBALS Array**



```
<?php
$y = 20;

function myFunction() {
    echo "Value inside function: " . $GLOBALS['y'];
}

myFunction(); // Output: Value inside function: 20
?>

```



# **4. Variables in Nested Functions**

A nested function is a function inside another function. The inner function cannot access local variables of the outer function unless passed explicitly.



```
<?php
function outerFunction() {
    $outerVar = "Hello";

    function innerFunction() {
        echo $outerVar; // ERROR: Undefined variable
    }

    innerFunction();
}
outerFunction();
?>

```



‚ùå Error: $outerVar is not accessible inside innerFunction().

# **‚úÖ Solution: Pass the variable as an argument:**



```
<?php
function outerFunction() {
    $outerVar = "Hello";

    function innerFunction($var) {
        echo $var;
    }

    innerFunction($outerVar); // Output: Hello
}
outerFunction();
?>

```



# **5. Example of Local & Global Variables**



```
<?php
$x = 50; // Global variable

function test() {
    global $x; // Access global variable inside function
    $y = 10; // Local variable
    echo "Global x: $x, Local y: $y";
}

test();
// echo $y; // ERROR: Undefined variable (Local variable)
?>

```



# **‚úÖ Output: Global x: 50, Local y: 10**

# **Include File in Another File in PHP**

# **1. Why Do We Need to Include Files?**

**Including files in PHP helps in:**

‚úî Code Reusability ‚Äì Avoids rewriting the same code in multiple files.

‚úî Easy Maintenance ‚Äì Updating a single file reflects changes in all places.

‚úî Better Organization ‚Äì Helps separate logic, HTML, and database configurations.

‚úî Modular Development ‚Äì Allows teamwork on different sections without conflicts.

# **2. include in PHP**

üîπ include is used to insert the contents of one file into another. If the file is missing, PHP shows a warning but continues execution.



```
<?php
include "header.php";
echo "Main content of the page.";
?>

```



üîπ If header.php exists ‚Üí It will be included.

üîπ If header.php does not exist ‚Üí PHP shows a warning but continues execution.

# **3. include_once in PHP**

üîπ include_once ensures a file is included only once to avoid redeclaration errors.



```
<?php
include_once "header.php";
include_once "header.php"; // This will not include again
?>

```



Screenshot 2025-03-31 093454.png

**üîπ Prevents duplicate inclusion and avoids issues like redeclared functions.**

# **4. require in PHP**

üîπ require works like include, but if the file is missing, PHP throws a fatal error and stops execution.



```
<?php
require "config.php";
echo "This will not run if config.php is missing.";
?>

```



**üîπ If config.php is missing ‚Üí PHP stops execution with a fatal error.**

# **5. require_once in PHP**

üîπ require_once works like require, but ensures the file is included only once to prevent multiple declarations.



```
<?php
require_once "config.php";
require_once "config.php"; // This will not include again
?>

```



**üîπ Ensures no duplicate inclusion.**

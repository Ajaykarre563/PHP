# **1. What is $_REQUEST?**

$_REQUEST is a PHP superglobal array that collects data sent via GET, POST, or COOKIE methods.

It is used to retrieve form data from an HTML form without specifying the method type.

# **Note**

**It is used for both get and post requests**

# **2. Make HTML and PHP Files**

You need an HTML file for the form and a PHP file to process the data.

The HTML file contains an input field and a submit button.

The PHP file retrieves the input data using $_REQUEST.

# **PHP File Upload Using $_FILES**

# **1. What is $_FILES?**

$_FILES is a PHP superglobal array used to collect information about uploaded files. It stores details like:

$_FILES['file']['name'] → Original file name

$_FILES['file']['type'] → File type (e.g., image/png)

$_FILES['file']['tmp_name'] → Temporary file location

$_FILES['file']['error'] → Upload error code

$_FILES['file']['size'] → File size in bytes

# **1. What are Cookies?**

A cookie is a small piece of data stored on the client's browser by a web server.

It is used to track user sessions, store preferences, and manage authentication.

# **2. Syntax and Parameters**

The setcookie() function is used to create cookies in PHP.




```
setcookie(name, value, expire, path, domain, secure, httponly);

```



name: Name of the cookie.

value: Value stored in the cookie.

expire: Expiry time in seconds (using time() + seconds).

path: Directory on the server where the cookie is available.

domain: Specifies the domain where the cookie is valid.

secure: If true, the cookie is sent over HTTPS only.

httponly: If true, the cookie is accessible only via HTTP (not JavaScript).

# **Set Data in Cookies**



```
setcookie("user", "JohnDoe", time() + (86400 * 30), "/"); // Expires in 30 days

```



# **Explanation:**

"user" → The name of the cookie.

"JohnDoe" → The value stored in the cookie.

time() + (86400 * 30) → Expiry time.

time() returns the current timestamp.

86400 is the number of seconds in one day.

86400 * 30 makes it expire in 30 days.

"/" → The cookie is available across the entire website.

# **Get Data from Cookies**



```
if(isset($_COOKIE["user"])) {
    echo "User: " . $_COOKIE["user"];
} else {
    echo "Cookie is not set.";
}

```

# **cookies example**



```
<form action="cookies.php" method="post">
    <input type="text" name="user" placeholder="Enter user name" />
    <br /><br />
    <button name="button" value="set">Set Cookies</button>
    <br /><br />
    <button name="button" value="display">Display Cookies</button>
    <br /><br />
    <button name="button" value="delete">Delete Cookies</button>
</form>

```





```
<?php
if ($_SERVER["REQUEST_METHOD"] == "POST") {
    if ($_POST['button'] == 'set') {
        // Set a cookie for 1 hour
        setcookie("user", $_POST['user'], time() + 3600, "/");
        echo "Cookie is set!";
    }

    if ($_POST['button'] == 'display') {
        // Display the cookie if it exists
        if (isset($_COOKIE['user'])) {
            echo "Stored Cookie Value: " . $_COOKIE['user'];
        } else {
            echo "No cookie found!";
        }
    }

    if ($_POST['button'] == 'delete') {
        // Delete the cookie
        if (isset($_COOKIE['user'])) {
            setcookie("user", "", time() - 3600, "/"); // Set expiration time in the past
            echo "Cookie has been deleted!";
        } else {
            echo "No cookie to delete!";
        }
    }
}
?>

```

# **Session in PHP**

# **1. What is a Session?**

A session in PHP is a way to store user information across multiple pages. Unlike cookies, session data is stored on the server, making it more secure.

# **2. Why Do We Use Sessions?**

To store user-specific data (e.g., login details, preferences) across multiple pages.

Unlike cookies, session data is not stored on the user's computer, reducing security risks.

Sessions expire after a certain period of inactivity, improving security.

Screenshot 2025-04-02 165914.png

# **4. Example**



```
<?php
session_start(); // Start session

// Store session data
$_SESSION["user"] = "Alice";
$_SESSION["email"] = "alice@example.com";

echo "Session variables are set.";
?>

```



**To access session data on another page:**



```
<?php
session_start();
echo "User: " . $_SESSION["user"];
echo "Email: " . $_SESSION["email"];
?>

```



# **SESSION example**



```
<!DOCTYPE html>
<html lang="en">
<head>
    <title>Sessions</title>
</head>
<body>
    <form action="" method="post">
        <input type="text" name="user" placeholder="Enter user name" />
        <br /><br />
        <button name="button" value="set">Set Session</button>
        <br /><br />
        <button name="button" value="get">Get Session</button>
        <br /><br />
        <button name="button" value="delete">Delete Session</button>
    </form>
</body>
</html>

<?php
session_start(); // Start the session

if ($_SERVER["REQUEST_METHOD"] == "POST") {
    if ($_POST['button'] == 'set') {
        // Set session variable
        $_SESSION["user"] = $_POST['user'];
        echo "Session is set for user: " . $_SESSION["user"];
    }

    if ($_POST['button'] == 'get') {
        // Display session variable
        if (isset($_SESSION["user"])) {
            echo "Stored Session Value: " . $_SESSION["user"];
        } else {
            echo "No session found!";
        }
    }

    if ($_POST['button'] == 'delete') {
        // Destroy the session
        session_unset();  // Unset all session variables
        session_destroy(); // Destroy the session
        echo "Session has been deleted!";
    }
}
?>

```
